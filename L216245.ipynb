Define a procedure called "check_input" that takes an input as a square list of lists (2D Matrix of N x N) and returns a boolen "true" if the input is valid or "false" otherwise, according to the following criteria.

1. Each coloumn must contain numbers from between 1 and N, with no duplicates.
2. Each row must contain numbers from between 1 and N, with no duplicates.

```python
# define procedure here
def check_input(data):
    n = len(data)
    status=True

    #Check each column in turn
    for column in range(0,n):
        # Start from the top of the column and move downwards to the second-to-last cell
        for row in range(0,n-1):
            # Temp variable may improve performance by hoisting row*column.
            thisCell = data[row][column]
            #Compare this cell to every cell below it in the same column.
            for checkRow in range(row+1,n):

                if (thisCell == data[checkRow][column]):

                #Found a duplicate
                    status=False
    return status
```

```python
# run this block to test code after defining the procedure

test1 = [[1, 2, 3],
        [2, 3, 1],
        [3, 1, 2]]

test2 = [[1, 2, 3, 4],
        [2, 3, 1, 3],
        [3, 1, 2, 3],
        [4, 4, 4, 4]]

test3 = [['a', 'b', 'c'],
        ['b', 'c', 'a'],
        ['c', 'a', 'b']]

print(check_input(test1)) # must return true
print(check_input(test2)) # must return false
print(check_input(test3)) # must return true
```
